```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df_train = pd.read_csv("train.csv")
df_train.head()
```

```python
df = pd.get_dummies(df_train, columns=['VIP','CryoSleep','Transported','HomePlanet'], drop_first=True) #one-hot encoding
df.drop(columns=['PassengerId','Name','Destination','Cabin'],inplace=True) 
```

```python
df.head()
```

```python
X = df.drop(columns=['Transported_True'])  #input
y = df['Transported_True']  #output
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) #80% for train and 20% for test
```

```python
from sklearn.preprocessing import StandardScaler #data scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(32,activation='relu',input_dim=10))
model.add(Dense(64,activation='linear'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(X_train_scaled, y_train, epochs=100)
```

```python
y_value = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_value>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)  
accuracy
```
